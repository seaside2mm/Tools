```python

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

##  1.1 线

```python
#############################################################################################################################
# 1. 如何画线的基础
#############################################################################################################################

x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2

# 基础部分
plt.figure(num=3, figsize=(8, 5),)
plt.plot(x, y2)
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')

#坐标轴的改动
plt.xlim((-1, 2))  #设置坐标范围
plt.ylim((-2, 3))
plt.xlabel('I am x')  #设置标签
plt.ylabel('I am y')

#可注释，比较变化
new_ticks = np.linspace(-1, 2, 5)
print("x轴刻度:",new_ticks)
plt.xticks(new_ticks)    #设置x轴刻度
plt.yticks([-2, -1.8, -1, 1.22, 3],[r'$really\ bad$', r'$bad$', r'$normal$', r'$good$', r'$really\ good$'])  #设置y轴刻度以及名称

plt.show()


#?plt.figure()
# plt.plot?  
```

```python
#############################################################################################################################
## 2. 设置坐标轴
#############################################################################################################################

x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2
plt.figure(num=3, figsize=(8, 5),)
plt.plot(x, y2)
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')
plt.xlim((-1, 2))  
plt.ylim((-2, 3))
plt.xlabel('I am x')  
plt.ylabel('I am y')
new_ticks = np.linspace(-1, 2, 5)
print(new_ticks)
plt.xticks(new_ticks)    
plt.yticks([-2, -1.8, -1, 1.22, 3],[r'$really\ bad$', r'$bad$', r'$normal$', r'$good$', r'$really\ good$'])  #设置y轴刻度以及名称

###############################################            开始             ########################################################

ax = plt.gca()  #获取当前坐标轴信息
ax.spines['right'].set_color('none')   # 设置边框
ax.spines['top'].set_color('none')    #使用.spines设置边框：最顶上的线条颜色消失；

#取消注释看看区别

#ax.xaxis.set_ticks_position('bottom')   #设置x坐标刻度数字或名称的位置, 包括top，bottom，both，default，none
#ax.spines['bottom'].set_position(('data', 0))  #使用.set_position设置边框位置：y=0的位置；（位置所有属性：outward，axes，data）

#ax.yaxis.set_ticks_position('left')
#ax.spines['left'].set_position(('data',0))

plt.show()
#?plt.gca
```

```python
#############################################################################################################################
## 3. 添加说明文字
#############################################################################################################################

x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2
plt.figure()
#set x limits
plt.xlim((-1, 2))
plt.ylim((-2, 3))
new_sticks = np.linspace(-1, 2, 5)
plt.xticks(new_sticks)
plt.yticks([-2, -1.8, -1, 1.22, 3],
           [r'$really\ bad$', r'$bad$', r'$normal$', r'$good$', r'$really\ good$'])

################################               开始                  ##############################################

l1, = plt.plot(x, y1, label='linear line')
l2, = plt.plot(x, y2, color='red', linewidth=1.0, linestyle='--', label='square line')

plt.legend(loc='upper right')  #显示的信息来自于上面代码中的 label

# 如果我们想单独修改之前的 label 信息, 给不同类型的线条设置图例信息. 我们可以在 plt.legend 输入更多参数. 
# 如果以下面这种形式添加 legend, 我们需要确保, 在上面的代码 plt.plot(x, y2, label='linear line') 和
# plt.plot(x, y1, label='square line') 中有用变量 l1 和 l2 分别存储起来. 
# 而且需要注意的是 l1, l2,要以逗号结尾, 因为plt.plot() 返回的是一个列表.
plt.legend(handles=[l1, l2], labels=['up', 'down'],  loc='best')

plt.show()
#?plt.legend()
```

```python
#画直线
x = np.linspace(-3, 3, 50)
y = 2*x + 1
plt.figure(num=1, figsize=(8, 5),)
plt.plot(x, y,)

#移动坐标
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

#############################################################################################################################
## 4. 标注
#############################################################################################################################
x0 = 1
y0 = 2*x0 + 1
plt.plot([x0, x0,], [0, y0,], 'k--', linewidth=2.5)    #标注出点(x0, y0)的位置信息，画出一条垂直于x轴的虚线
# set dot styles
plt.scatter([x0, ], [y0, ], s=50, color='b')

#添加说明信息
plt.annotate(r'$2x+1=%s$' % y0, xy=(x0, y0), xycoords='data', xytext=(+30, -30),  #参数xycoords='data' 是说基于数据的值来选位置, 
             textcoords='offset points', fontsize=16,     #xytext=(+30, -30) 和 textcoords='offset points' 对于标注位置的描述 和 xy 偏差值, 
             arrowprops=dict(arrowstyle='->', connectionstyle="arc3,rad=.2"))    #arrowprops是对图中箭头类型的一些设置.

#添加注释
plt.text(-3.7, 3, r'$This\ is\ the\ some\ text. \mu\ \sigma_i\ \alpha_t$',   #其中-3.7, 3,是选取text的位置, 
         fontdict={'size': 16, 'color': 'r'})                #空格需要用到转字符\ ,fontdict设置文本字体.

plt.show()

#?plt.annotate
```

#  1.3 等高线

```python
#创建数据集
def f(x,y):
    # the height function
    return (1 - x / 2 + x**5 + y**3) * np.exp(-x**2 -y**2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X,Y = np.meshgrid(x, y)         #在二维平面中将每一个x和每一个y分别对应起来，编织成栅格

# use plt.contourf to filling contours
# X, Y and value for (X,Y) point
plt.contourf(X, Y, f(X, Y), 8, alpha=.75, cmap=plt.cm.hot)    #将 f(X,Y) 的值对应到color map的暖色组中寻找对应颜色

C = plt.contour(X, Y, f(X, Y), 8, colors='black', linewidth=.5)   #等高线绘制

plt.clabel(C, inline=True, fontsize=10)  # 添加高度数字, inline控制是否将Label画在线里面
plt.xticks(())        #坐标轴隐藏
plt.yticks(())

plt.show()
#?np.meshgrid()
#?plt.contourf()
```

# 1.4  3D图像

```python
fig = plt.figure()
ax = Axes3D(fig)  #创建3d窗口

# X, Y value
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)    # x-y 平面的网格
R = np.sqrt(X ** 2 + Y ** 2)
# height value
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.get_cmap('rainbow'))  #rstride 和 cstride 分别代表 row 和 column 的跨度

#ax.contourf(X, Y, Z, zdir='z', offset=-2, cmap=plt.get_cmap('rainbow'))  #投影

fig.show()
```
